```python
def remover(textos, regex):
  if type(textos) == str:
    return regex.sub("", textos)
  else:
    return [regex.sub("", texto) for texto in textos]
```

```python
regex_html = re.compile(r"<.*?>")

questao_sem_tag = remover(questao_ingles, regex_html)
print(questao_sem_tag)
```

###Detectando e removendo os códigos das questões. É uma boa prática remover os códigos para que o modelo não considere partes dos códigos como texto.

```python
def substituir_codigo(textos, regex):
  if type(textos) == str:
    return regex.sub("CODE", textos)
  else:
    return [regex.sub("CODE", texto) for texto in textos]
```

```python
regex_codigo = re.compile(r"<code>(.|(\n))*?</code>")
```

```python
questao_sem_code = substituir_codigo(questao_espanhol, regex_codigo)
print(questao_sem_code)
```

```python
from sklearn.model_selection import train_test_split

port_treino, port_teste = train_test_split(df_port.questoes_tratadas, test_size = 0.2, random_state = 123)
esp_treino, esp_teste = train_test_split(df_esp.questoes_tratadas, test_size = 0.2, random_state = 123)
ing_treino, ing_teste = train_test_split(df_ing.questoes_tratadas, test_size = 0.2, random_state = 123)
```

###Todas as questões em uma variavel
- Juntando todas as questões em uma única variável

```python
todas_questoes_port = ' '.join(port_treino)
todas_questoes_esp = ' '.join(esp_treino)
todas_questoes_ing = ' '.join(ing_treino)
```

###Separando palavra por palavra
- **WhitespaceTokenizer:** É uma classe do NLTK que tokeniza (ou separa) um texto com base em espaços em branco. Isso significa que o texto será dividido em palavras onde quer que haja um espaço em branco.

- O código está dividindo três strings de texto (todas_questoes_port, todas_questoes_esp, todas_questoes_ing) em listas de palavras, baseando-se em espaços em branco como delimitadores.

```python
from nltk.tokenize import WhitespaceTokenizer

todas_palavras_port = WhitespaceTokenizer().tokenize(todas_questoes_port)
todas_palavras_esp = WhitespaceTokenizer().tokenize(todas_questoes_esp)
todas_palavras_ing = WhitespaceTokenizer().tokenize(todas_questoes_ing)
```

- **padded_everygram_pipeline:** Uma função que cria n-gramas e vocabulário com padding (preenchimento) para treinamento de modelos de linguagem.

- **A função padded_everygram_pipeline realiza duas tarefas principais:**

- **Criação de N-Gramas com Padding:**
Gera n-gramas (no caso, bigramas) a partir da lista de palavras.
Adiciona padding (preenchimento) no início e no fim das sequências para capturar contextos de início e fim de sentença. Isso ajuda na modelagem de linguagem, pois permite que o modelo aprenda a prever palavras no início e no fim das sentenças.
- **Criação de Vocabulário:**
Gera um vocabulário a partir da lista de palavras, que é utilizado pelo modelo de linguagem para conhecer todas as palavras possíveis que pode encontrar.

```python
from nltk.lm.preprocessing import padded_everygram_pipeline

port_treino_bigram, vocab_port = padded_everygram_pipeline(2, todas_palavras_port)
esp_treino_bigram, vocab_esp = padded_everygram_pipeline(2, todas_palavras_esp)
ing_treino_bigram, vocab_ing = padded_everygram_pipeline(2, todas_palavras_ing)
```

###Aplicando modelo MLE
- A Estimação de Máxima Verossimilhança (MLE, Maximum Likelihood Estimation) é um método estatístico utilizado para estimar os parâmetros de um modelo probabilístico. No contexto de modelos de linguagem n-grama, a MLE é utilizada para calcular as probabilidades de sequências de palavras com base nas frequências observadas nos dados de treinamento.

```python
from nltk.lm import MLE

modelo_port = MLE(2) #Bigrama
modelo_esp = MLE(2)
modelo_ing = MLE(2)

modelo_port.fit(port_treino_bigram, vocab_port)
modelo_esp.fit(esp_treino_bigram, vocab_esp)
modelo_ing.fit(ing_treino_bigram, vocab_ing)
```

```python
modelo_port.generate(num_words = 6)
```

```python
def treinar_modelo_mle(lista_textos):
  todas_questoes = ' '.join(lista_textos)
  todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
  bigrams, vocabulario = padded_everygram_pipeline(2, todas_palavras)
  modelo = MLE(2)
  modelo.fit(bigrams, vocabulario)
  return modelo
```

```python
modelo_port_2 = treinar_modelo_mle(port_treino)
```

```python
modelo_ing = treinar_modelo_mle(ing_treino)
modelo_esp = treinar_modelo_mle(esp_treino)
```

```python
#Função para calcular a perplexidade dos textos
def calcular_perplexidade(modelo, texto):
  perplexidade = 0
  palavras = WhitespaceTokenizer().tokenize(texto)
  palavras_fakechar = [list(pad_both_ends(palavra, n = 2))for palavra in palavras]
  palavras_bigramns = [list(bigrams(palavra)) for palavra in palavras_fakechar]

  for palavra in palavras_bigramns:
    perplexidade += modelo.perplexity(palavra)
  return perplexidade
```

 ### Técnica de Laplace faz com que as probabilidades nunca seja 0, isso evita que o calculo seja infinito

```python
from nltk.lm import Laplace
def treinar_modelo_Laplace(lista_textos):
  todas_questoes = ' '.join(lista_textos)
  todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
  bigrams, vocabulario = padded_everygram_pipeline(2, todas_palavras)
  modelo = Laplace(2)
  modelo.fit(bigrams, vocabulario)
  return modelo
```

```python
modelo_ing_laplace = treinar_modelo_Laplace(ing_treino)
print(calcular_perplexidade(modelo_ing_laplace, port_teste.iloc[0]))
```

```python
modelo_port_laplace = treinar_modelo_Laplace(port_treino)
print(calcular_perplexidade(modelo_port_laplace, port_teste.iloc[0]))
```

```python
modelo_esp_laplace = treinar_modelo_Laplace(esp_treino)
print(calcular_perplexidade(modelo_esp_laplace, port_teste.iloc[0]))
```

### O modelo com menor perplexidade é o que mais se ajusta. Como estou fazendo o teste do modelo com a base de Português, o modelo com menor perplexidade será o que tem os dados de treino em português

###Função para atribuir idioma

```python
def atribuir_idioma(lista_textos):
  idioma = []
  for texto in lista_textos:
    portugues = calcular_perplexidade(modelo_port_laplace, texto)
    espanhol = calcular_perplexidade(modelo_esp_laplace, texto)
    ingles = calcular_perplexidade(modelo_ing_laplace, texto)
    if ingles >= portugues <= espanhol:
      idioma.append("portugues")
    elif portugues > ingles < espanhol:
      idioma.append('ingles')
    else:
      idioma.append('espanhol')
  return idioma
```

```python
resultados_port = atribuir_idioma(port_teste)
resultados_esp = atribuir_idioma(esp_teste)
resultados_ing = atribuir_idioma(ing_teste)
```
